### Import packages 

```python
import sqlite3
import pandas as pd
```

### Connect to database

```python
conn = sqlite3.connect('amazon_db.db')
c = conn.cursor()
```

```python
c.execute("""SELECT * FROM books
LIMIT 5; """) 
```

```python
df = pd.DataFrame(c.fetchall())
df.columns = [i[0] for i in c.description]
df
```

```python
#How many rows are in the 'author' database?
c.execute("""SELECT * FROM author LIMIT 5;""")
df = pd.DataFrame(c.fetchall())
df.columns = [i[0] for i in c.description]
df
```

```python
#How many authors are called David?
c.execute("""SELECT count(*) 
            FROM author WHERE author LIKE "David%";""")

df = pd.DataFrame(c.fetchall())
df.columns = [i[0] for i in c.description]
df
```

```python
#How many authors are called David?
c.execute("""SELECT author.author, sum(books.num_copies) 
             FROM author
             LEFT JOIN books
                   ON author.book = books.books
             GROUP BY author
             ORDER BY count(author.book) desc;""")

df = pd.DataFrame(c.fetchall())
df.columns = [i[0] for i in c.description]
df
```

```python
#Use both the author table and books table to say: how many copies were sold by authors named David
#How many authors are called David?
c.execute("""SELECT author.author, count(author.book) 
             FROM author
             LEFT JOIN books
                   ON author.book = books.books
             WHERE author.author Like "David%"
             GROUP BY author
             ORDER BY count(author.book) desc;""")

df = pd.DataFrame(c.fetchall())
df.columns = [i[0] for i in c.description]
df
```

```python
#b. how many books in total for all Davids
c.execute("""SELECT sum(books.num_copies) 
             FROM author
             LEFT JOIN books
                   ON author.book = books.books
             WHERE author.author Like "David%"
             ORDER BY count(author.book) desc;""")

df = pd.DataFrame(c.fetchall())
df.columns = [i[0] for i in c.description]
df
```
