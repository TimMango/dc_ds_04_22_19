4. Write a function that calculates the chances to get 60 heads or more in 100 tosses of a coin if the coin is fair. 

Hint: Do this by running a simulation.

```python
import numpy as np

###To make list of coin flips
sample_size = 1000000
sixty_heads=[]
under_sixty=[]

for i in range(sample_size):
    #100 coin flips with a probability of flipping heads at 0.5
    #sample size is 1
    coin_flip = np.random.binomial(100, .5, 1)
    number_heads = sum(coin_flip)
    if number_heads > 60:
        sixty_heads.append(number_heads)
    else:
        under_sixty.append(number_heads)
##Proportion of heads greater than 60
probability = len(sixty_heads)/(len(sixty_heads)+len(under_sixty))
print(probability)

#Probability is roughly 1.75%
```
